```python
def count_paragraphs(text):
    count = 0
    for para in text.split("\n"):
        count += 1
    return count

def get_paragraphs(text):
    para_counts = {}
    count = 0
    for para in text.split("\n"):
        count += 1
        para_counts[count] = para + '\n======================\n'
    return para_counts

def name_paragraphs(text):
    para_counts = {}
    count = 0
    for para in text.split("\n"):
        count += 1
        para_counts[count] = "Paragraph " + count
    return para_counts

def get_sentences(text):
    sent_counts = {}
    count = 0
    for sent in text.split(". "):
        count += 1
        sent_counts[count] = sent + '\n======================\n'
    return sent_counts

def count_sentences(text):
    count = 0
    for sent in text.split(". "):
        count += 1
    return count

def count_words(text):
    skips = [".", ", ", ":", ";", "'", '"']
    for ch in skips:
        text = text.replace(ch, "")
    word_counts = {}
    for word in text.split(" "):
        if word in word_counts:
            word_counts[word]+= 1
        else:
            word_counts[word]= 1
    return word_counts
```
